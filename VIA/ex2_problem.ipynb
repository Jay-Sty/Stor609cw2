# Example 2 work through

This is a work through of example 2 from exercise 9.27 in section 9.5 of Artificial Intelligence: Foundations and Computational Agents 2nd edition. We shall see the problem and how to solve it manually and with code using the value iterated algorithm.

## Problem Description - Sam's weekend

Sam can either rest or party at the weekend. He prefers to party but gains more from partying when healthy, and less when he is sick. The full problem is described as an MDP below:

| State  | Action | Probability of s' = healthy |
|:------:|:------:|:---------------------------:|
| healthy | relax  | 0.95                        |
| healthy | party  | 0.7                         |
| sick    | relax  | 0.5                         |
| sick    | party  | 0.1                         |

| State   | Action | Reward |
|:-------:|:------:|:------:|
| healthy | relax  | 7      |
| healthy | party  | 10     |
| sick    | relax  | 0      |
| sick    | party  | 2      |

## Sam's Weekend Example – Value Iteration (2 Iterations)

**States:** $S = \{\text{healthy}, \text{sick}\}$  
**Actions:** $A = \{\text{relax}, \text{party}\}$  
**Discount factor:** $\gamma = 0.9$  

**Transition probabilities:**


\begin{array}{c c c c}
\text{State} & \text{Action} & P(s' = \text{healthy}) & P(s' = \text{sick}) \\
\hline
\text{healthy} & \text{relax} & 0.95 & 0.05 \\
\text{healthy} & \text{party} & 0.7 & 0.3 \\
\text{sick} & \text{relax} & 0.5 & 0.5 \\
\text{sick} & \text{party} & 0.1 & 0.9
\end{array}


**Rewards:**


\begin{array}{c c c}
\text{State} & \text{Action} & R(s,a) \\
\hline
\text{healthy} & \text{relax} & 7 \\
\text{healthy} & \text{party} & 10 \\
\text{sick} & \text{relax} & 0 \\
\text{sick} & \text{party} & 2
\end{array}




### Step 0: Initialization


$V_0(\text{healthy}) = 0, \quad V_0(\text{sick}) = 0$



### Step 1: First Iteration $(k=1)$

Value iteration formula:


$V_{k+1}(s) = \max_a \sum_{s'} P(s'|s,a) \big[ R(s,a) + \gamma V_k(s') \big]$


Since $V_0 = 0$, only immediate rewards matter.

**Healthy state:**


\begin{aligned}
V_1(\text{healthy, relax}) &= 0.95 \cdot 7 + 0.05 \cdot 7 = 7 \\
V_1(\text{healthy, party}) &= 0.7 \cdot 10 + 0.3 \cdot 10 = 10 \\
V_1(\text{healthy}) &= \max(7, 10) = 10 \quad \Rightarrow \text{Policy: party}
\end{aligned}


**Sick state:**


\begin{aligned}
V_1(\text{sick, relax}) &= 0.5 \cdot 0 + 0.5 \cdot 0 = 0 \\
V_1(\text{sick, party}) &= 0.1 \cdot 2 + 0.9 \cdot 2 = 2 \\
V_1(\text{sick}) &= \max(0, 2) = 2 \quad \Rightarrow \text{Policy: party}
\end{aligned}




### Step 2: Second Iteration $(k=2)$

Now include discounted future values:


$V_2(s) = \max_a \sum_{s'} P(s'|s,a) \big[ R(s,a) + \gamma V_1(s') \big]$


**Healthy state:**


\begin{aligned}
V_2(\text{healthy, relax}) &= 0.95 \cdot (7 + 0.9 \cdot 10) + 0.05 \cdot (7 + 0.9 \cdot 2) \\
&= 0.95 \cdot 16 + 0.05 \cdot 8.8 \\
&= 15.2 + 0.44 = 15.64 \\
V_2(\text{healthy, party}) &= 0.7 \cdot (10 + 0.9 \cdot 10) + 0.3 \cdot (10 + 0.9 \cdot 2) \\
&= 0.7 \cdot 19 + 0.3 \cdot 11.8 \\
&= 13.3 + 3.54 = 16.84 \\
V_2(\text{healthy}) &= \max(15.64, 16.84) = 16.84 \quad \Rightarrow \text{Policy: party}
\end{aligned}


**Sick state:**


\begin{aligned}
V_2(\text{sick, relax}) &= 0.5 \cdot (0 + 0.9 \cdot 10) + 0.5 \cdot (0 + 0.9 \cdot 2) \\
&= 0.5 \cdot 9 + 0.5 \cdot 1.8 \\
&= 4.5 + 0.9 = 5.4 \\
V_2(\text{sick, party}) &= 0.1 \cdot (2 + 0.9 \cdot 10) + 0.9 \cdot (2 + 0.9 \cdot 2) \\
&= 0.1 \cdot 11 + 0.9 \cdot 3.8 \\
&= 1.1 + 3.42 = 4.52 \\
V_2(\text{sick}) &= \max(5.4, 4.52) = 5.4 \quad \Rightarrow \text{Policy: relax}
\end{aligned}




### Values after 2 Iterations


\begin{array}{c c c}
\text{State} & V_2(s) & \text{Policy} \\
\hline
\text{healthy} & 16.84 & \text{party} \\
\text{sick} & 5.4 & \text{relax}
\end{array}


**Summary:**  

- Emerging optimal policy: `healthy → party`, `sick → relax`  
- Values are the expected total discounted reward starting from each state.

## Value iteration Algorithm Code Result

Can also compare with the algorthm coded:

```python
from VIA import value_iteration

##example 2 (ex9.27 from of Artificial Intelligence: Foundations and Computational Agents 2nd edition)


def sam_weekend_mdp():
    states = ['healthy', 'sick']
    actions = ['relax', 'party']

    #Transition and reward dictionaries
    P_dict = {
        'healthy': {'relax': {'healthy': 0.95, 'sick': 0.05},
                    'party': {'healthy': 0.7, 'sick': 0.3}},
        'sick': {'relax': {'healthy': 0.5, 'sick': 0.5},
                 'party': {'healthy': 0.1, 'sick': 0.9}}
    }

    R_dict = {
        'healthy': {'relax': 7, 'party': 10},
        'sick': {'relax': 0, 'party': 2}
    }

    #Wrapper functions
    def P(s, a, s_next):
        return P_dict[s][a].get(s_next, 0)

    def R(s, a, s_next):
        return R_dict[s][a]

    gamma = 0.9
    epsilon = 0.05
    max_iterations = 10000

    policy, value_function, delta_list = value_iteration(states, actions, P, R, gamma, epsilon, max_iterations)
    return policy, value_function

if __name__ == "__main__":
    #only run if script run directly (not imported)
    policy, value_function = sam_weekend_mdp()
    print("Optimal Policy:", policy)
    print("Value Function:", value_function)
```
